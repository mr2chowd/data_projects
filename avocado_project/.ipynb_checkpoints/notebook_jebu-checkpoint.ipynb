Repeat the steps above for the rest of the ingredients. To make this replicable, you can create a function called read_and_filter_data() 
which takes two arguments, filepath and relevant_categories, and returns a filtered DataFrame 
containing only rows where categories_tags are in a list called relevant_categories. The function also prints the number of rows per origin, 
 which you'll use to identify the most likely country of origin.

```python
def read_and_filter_data(filepath, relevant_categories):
  df = pd.read_csv(filepath, sep='\t')

  # Subset data
  df = df[subset_columns]

  # Split tags into lists
  df['categories_list'] = df['categories_tags'].str.split(',')

  # Drop null categories and filter data
  df = df.dropna(subset = 'categories_list')

  df = df[df['categories_list'].apply(lambda x: any([i for i in x if i in relevant_categories]))]
    
  df = df[(df['countries']=='United Kingdom')]
  print(f'**{filepath[:-4]} origins**','\n',df['origins_tags'].value_counts(), '\n')

  return df
    
    
    
```
